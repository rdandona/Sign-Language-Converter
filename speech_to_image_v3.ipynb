```python
import speech_recognition as sr
import sys
import glob
import cv2
import string
from tkinter import *
from functools import partial
import tkinter.font
from PIL import ImageTk,Image
import time
```

```python
def speak():
    label1=Label(root,text='PROCESSING TEXT MAY TAKE SOME TIME',fg='blue',font=tk_font,bg='#fff')
    label1.place(relx=.5,rely=.19,anchor=CENTER)
    r=sr.Recognizer()
    with sr.Microphone(device_index=8) as source:
        #saving the voice input in variable audio
        audio=r.listen(source,timeout=5,phrase_time_limit=5)
        label2=Label(root,text='PROCESSING...',fg='blue',font=tk_font,bg='#fff')
        label2.place(relx=.5,rely=.25,anchor=CENTER)
        
        try:
#             inputDialogue=r.recognize_google(audio, key="GOOGLE_SPEECH_RECOGNITION_API_KEY")
#             label2.config(text='You Said: '+inputDialogue)
            inputDialogue='good morning how are you'
            process_text(inputDialogue)
        except:
            label2.config(text='COULD NOT RECOGNIZE')
            
        
        
    
        

def process_text(inputDialogue):
    #splitting the sentence to list of words
    wordsSpoken=inputDialogue.split()
    
    #to find the number of words spoken
    numberOfWordsSpoken=len(wordsSpoken)
    
    #conversion of uppercase letters to lower case
    for i in range(numberOfWordsSpoken):
        wordsSpoken[i]=wordsSpoken[i].lower()
        
    
                
    global flag 
    
    def multiple_img(flag,img):
        img.destroy()
        flag=flag+1
#         for i in range(flag,numberOfWordsSpoken):
        for words in glob.glob('./images/*'):
            actualWord=words.split("/")[2].split(".")[0]
            if actualWord !=wordsSpoken[flag]:
                continue
            else:
                load = Image.open(words)
                render = ImageTk.PhotoImage(load)
                img = Label(root, image=render)
                img.image = render
                img.place(relx=.5, rely=.5, anchor=CENTER)
                

#         break
    
    #iterating through the database 
    #if a match is found, the relevent image is displayed
    
    if numberOfWordsSpoken==1:
        for words in glob.glob('./images/*'):
            actualWord=words.split("/")[2].split(".")[0]
            if actualWord !=wordsSpoken[0]:
                 continue
            else:
                load = Image.open(words)
                render = ImageTk.PhotoImage(load)
                img = Label(root, image=render)
                img.image = render
                img.place(relx=.5, rely=.5, anchor=CENTER)
    else:
        flag=0
        for words in glob.glob('./images/*'):
            actualWord=words.split("/")[2].split(".")[0]
            if actualWord !=wordsSpoken[0]:
                 continue
            else:
                load = Image.open(words)
                render = ImageTk.PhotoImage(load)
                img = Label(root, image=render)
                img.image = render
                img.place(relx=.5, rely=.5, anchor=CENTER)
                
                
                break
                
    multiple_img=partial(multiple_img,flag,img)
    next_btn=Button(root,text='NEXT',font=tk_font,width=25,height=3,bg='#42f498',fg='#fff',command=multiple_img)
    next_btn.place(relx=.5, rely=.9, anchor=CENTER) 
            
               

root = Tk()
root.geometry('1000x800+0+0')
root.title('S2I')
root.config(bg="#ffffff")

tk_font=tkinter.font.Font(root=root,family='Times',size='16')
        
label1=Label(root,text='PROCESSING TEXT MAY TAKE SOME TIME',fg='blue',font=tk_font,bg='#fff')
label1.place(relx=.5,rely=.19,anchor=CENTER)

speak_btn=Button(root,text='SPEAK',font=tk_font,width=25,height=3,bg='#42f498',fg='#fff',command=speak)
speak_btn.place(relx=.5, rely=.1, anchor=CENTER)



root.mainloop()
```
